###### Titanic przykłady ML

```python
import pandas as pd

from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv('../day2/input/train.csv')
df_test = pd.read_csv('../day2/input/test.csv')
```

```python
df_train.sample(10)
```

```python
# tworzenie macieży
X = df_train[["Pclass"]].values
# zmienna y wektorowa
y = df_train["Survived"].values 
```

```python
# tworzenie modelu

model = DummyClassifier()
model.fit(X, y)
y_pred = model.predict(X)

score = accuracy_score(y, y_pred)
print("score: %.2f" % score)
```

```python

X = df_test[['Pclass']].values

df_test['Survived'] = model.predict(X)
df_test[ ['PassengerId', 'Survived'] ].to_csv('../day2/output/dummy.csv', index=False)
```
